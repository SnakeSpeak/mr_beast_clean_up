```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# create beast_df
beast = pd.read_csv("MrBeast_youtube_stats.csv")
beast_df = pd.DataFrame(beast)
beast_df.head()
```

```python
# info about beast_df
print(beast_df.columns)

print(beast_df.dtypes)

print(beast_df.value_counts)
```

```python
# turn 'publishTime' column into usable data
beast_df['publishTime'] = pd.to_datetime(beast_df['publishTime'])
```

```python
# check dtype of columns, 'publishTime' should be changed to datetime
beast_df.dtypes
```

```python
# repeat for pull_date
beast_df['pull_date'] = pd.to_datetime(beast_df['pull_date'])
```

```python
# check dtype of columns, 'pull_date' should be changed to datetime
beast_df.dtypes
```

```python
# create a DataFrame where it's only unique ids AND is the most recent pull_date
idx = beast_df.groupby('id')['pull_date'].idxmax()
recent_pull_beast_df = beast_df.loc[idx]
recent_pull_beast_df
```

```python
# set index to id and sort by publishTime
recent_pull_beast_df = recent_pull_beast_df.sort_values('publishTime')
```

```python
recent_pull_beast_df
```

```python
# clean up recent_pull_beast_df to only include relevant columns for analysis
beast_clean_df = recent_pull_beast_df.drop(['kind_stats','thumbnails.default.url', 'thumbnails.default.width',\
       'thumbnails.default.height', 'thumbnails.medium.url',\
       'thumbnails.medium.width', 'thumbnails.medium.height',\
       'thumbnails.high.url', 'thumbnails.high.width',\
       'thumbnails.high.height', 'contentDetails.duration',\
       'contentDetails.dimension', 'topicDetails.topicCategories',\
       'snippet.defaultLanguage','localizations.en.description',\
       'localizations.en.title','snippet.tags'], axis = 1)
print(beast_clean_df.head())
print(beast_clean_df.value_counts)
```

```python
beast_clean_df = beast_clean_df.reset_index()
beast_clean_df = beast_clean_df.drop('index',axis = 1)
```

```python
# create cleaned_beast.csv to use in Tableau
beast_clean_df.to_csv('clean_beast.csv')
```
